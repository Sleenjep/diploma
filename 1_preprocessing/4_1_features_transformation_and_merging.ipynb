```python
merge_enabled = True

exception_groups = {
    "Ступни (размер)",
    "Шея (размер)",
    "Волосы или шапка (аккуратность)",
    "Левый глаз (форма)",
    "Лоб (размер)",
    "Ноги (постановка)",
    "Правая нога (положение)",
    "Левая нога (положение)",
}

merge_analysis = {}
mergeable_groups = []
not_mergeable_groups = []
exception_rows_to_drop = set()

for new_col, source_map in features_to_merge.items():
    source_cols = list(source_map.keys())
    missing_cols = [c for c in source_cols if c not in df.columns]

    if missing_cols:
        merge_analysis[new_col] = {
            "status": "missing_columns",
            "missing_columns": missing_cols,
            "source_columns": source_cols,
        }
        continue

    filled_counts = df[source_cols].notna().sum(axis=1)
    conflict_mask = filled_counts > 1
    conflict_rows = int(conflict_mask.sum())

    if conflict_rows > 0:
        if new_col in exception_groups:
            merge_analysis[new_col] = {
                "status": "exception_can_merge",
                "source_columns": source_cols,
                "conflict_rows": conflict_rows,
                "rows_to_drop": df.index[conflict_mask].tolist(),
            }
            mergeable_groups.append(new_col)
            exception_rows_to_drop.update(df.index[conflict_mask].tolist())
        else:
            merge_analysis[new_col] = {
                "status": "cannot_merge",
                "source_columns": source_cols,
                "conflict_rows": conflict_rows,
            }
            not_mergeable_groups.append(new_col)
    else:
        merge_analysis[new_col] = {
            "status": "can_merge",
            "source_columns": source_cols,
        }
        mergeable_groups.append(new_col)

print("Можно объединить:")
for group_name in mergeable_groups:
    status = merge_analysis[group_name]["status"]
    if status == "exception_can_merge":
        conflict_rows = merge_analysis[group_name]["conflict_rows"]
        print(f"- {group_name} (исключение, будет удалено строк: {conflict_rows})")
    else:
        print(f"- {group_name}")

print("\nНельзя объединить:")
for group_name in not_mergeable_groups:
    conflict_rows = merge_analysis[group_name]["conflict_rows"]
    print(f"- {group_name} (строк с 2+ заполненными признаками: {conflict_rows})")

print("\nИсключения:")
for group_name in exception_groups:
    if group_name in merge_analysis and merge_analysis[group_name]["status"] == "exception_can_merge":
        conflict_rows = merge_analysis[group_name]["conflict_rows"]
        print(f"- {group_name}: будет удалено строк {conflict_rows}")

print("\nПропущены (нет нужных столбцов в df):")
for group_name, info in merge_analysis.items():
    if info["status"] == "missing_columns":
        print(f"- {group_name}: {info['missing_columns']}")

if merge_enabled:
    if exception_rows_to_drop:
        before_drop_shape = df.shape
        df = df.drop(index=sorted(exception_rows_to_drop)).reset_index(drop=True)
        print(f"\nУдалены строки по исключениям: {len(exception_rows_to_drop)}")
        print(f"Форма df после удаления строк: {before_drop_shape} -> {df.shape}")

    for new_col in mergeable_groups:
        source_map = features_to_merge[new_col]
        source_cols = list(source_map.keys())

        merged_series = pd.Series(np.nan, index=df.index, dtype=object)
        for source_col, label in source_map.items():
            mask = df[source_col].notna() & merged_series.isna()
            if mask.any():
                merged_series.loc[mask] = label

        df[new_col] = merged_series
        df = df.drop(columns=source_cols)

    print(f"\nОбъединение выполнено. Новая форма df: {df.shape}")
else:
    print("\nФлаг merge_enabled=False, объединение не выполнялось.")
```

```python
label_merge_report = {}

for feature_name, merge_rules in labels_to_merge.items():
    if feature_name not in df.columns:
        label_merge_report[feature_name] = {
            "status": "missing_feature",
            "changed_rows": 0,
        }
        continue

    if not merge_rules:
        label_merge_report[feature_name] = {
            "status": "empty_rules",
            "changed_rows": 0,
        }
        continue

    value_to_target = {}
    for target_label, source_values in merge_rules.items():
        for value in source_values:
            value_to_target[value] = target_label

    before_values = df[feature_name].copy()
    df[feature_name] = df[feature_name].map(lambda x: value_to_target.get(x, x) if pd.notna(x) else x)

    changed_mask = (before_values != df[feature_name]) & ~(before_values.isna() & df[feature_name].isna())
    changed_rows = int(changed_mask.sum())

    label_merge_report[feature_name] = {
        "status": "merged",
        "changed_rows": changed_rows,
    }

print("Результаты объединения значений:")
for feature_name, info in label_merge_report.items():
    if info["status"] == "merged":
        print(f"- {feature_name}: изменено строк {info['changed_rows']}")
    elif info["status"] == "missing_feature":
        print(f"- {feature_name}: признак отсутствует в df")
    elif info["status"] == "empty_rules":
        print(f"- {feature_name}: правила объединения не заданы")
```

```python
statistics = []

for col_idx in range(1, len(df.columns)):
    col_name = df.columns[col_idx]
    column_data = df[col_name]
    
    missing_count = int(column_data.isna().sum())
    
    value_counts = column_data.value_counts(dropna=False)
    
    stat_row = {
        'features': col_name,
        'unique_values': [],
        'frequencies': [],
        'missing_values': missing_count
    }
    
    has_empty = False
    for value, count in value_counts.items():
        value_str = str(value)
        if pd.isna(value) or value_str.lower() == 'nan':
            value_str = 'пусто'
            has_empty = True
        stat_row['unique_values'].append(value_str)
        stat_row['frequencies'].append(int(count))

    if missing_count > 0 and not has_empty:
        stat_row['unique_values'].append('пусто')
        stat_row['frequencies'].append(missing_count)

    statistics.append(stat_row)
```

```python
analysis_dir = f"./{prefix}_analysis"
os.makedirs(analysis_dir, exist_ok=True)

output_file = f"{analysis_dir}/{prefix}_statistics_analysis.xlsx"

max_unique_values = max(len(stat['unique_values']) for stat in statistics)

wb = Workbook()
ws = wb.active
ws.title = "Статистика признаков"

headers = ['Признак']
for i in range(max_unique_values):
    headers.append(f'Значение_{i+1}')
    headers.append(f'Частота_{i+1}')
headers.append('Пропуски')

cell_font = Font(bold=False)
cell_align = Alignment(horizontal='left', vertical='center')

for col_idx, header in enumerate(headers, 1):
    cell = ws.cell(row=1, column=col_idx, value=header)
    cell.font = cell_font
    cell.alignment = cell_align

for row_idx, stat in enumerate(statistics, 2):
    col_idx = 1
    
    c = ws.cell(row=row_idx, column=col_idx, value=stat['features'])
    c.font = cell_font
    c.alignment = cell_align
    col_idx += 1
    
    unique_values = stat['unique_values']
    frequencies = stat['frequencies']
    
    for i in range(max_unique_values):
        if i < len(unique_values):
            uv = unique_values[i]
            if isinstance(uv, str) and uv != 'пусто':
                try:
                    f = float(uv)
                    if f.is_integer():
                        uv = int(f)
                except ValueError:
                    pass
            c = ws.cell(row=row_idx, column=col_idx, value=uv)
            c.font = cell_font
            c.alignment = cell_align
            col_idx += 1
            c = ws.cell(row=row_idx, column=col_idx, value=int(frequencies[i]))
            c.font = cell_font
            c.alignment = cell_align
            col_idx += 1
        else:
            col_idx += 2
    
    c = ws.cell(row=row_idx, column=col_idx, value=int(stat['missing_values']))
    c.font = cell_font
    c.alignment = cell_align

for col in ws.columns:
    max_length = 0
    col_letter = col[0].column_letter
    for cell in col:
        try:
            if cell.value:
                max_length = max(max_length, len(str(cell.value)))
        except:
            pass
    adjusted_width = min(max_length + 2, 50)
    ws.column_dimensions[col_letter].width = adjusted_width

wb.save(output_file)
```

```python
reports_dir = os.path.join(analysis_dir, f"{prefix}_reports")
os.makedirs(reports_dir, exist_ok=True)

output_file = f"{reports_dir}/{prefix}_statistics_analysis.xlsx"

plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

rows_per_page = 4
cols_per_page = 4
graphs_per_page = rows_per_page * cols_per_page

total_features = len(statistics)
num_pages = math.ceil(total_features / graphs_per_page)

for page in range(num_pages):
    start_idx = page * graphs_per_page
    end_idx = min(start_idx + graphs_per_page, total_features)
    page_features = statistics[start_idx:end_idx]
    
    fig, axes = plt.subplots(rows_per_page, cols_per_page, figsize=(20, 16))
    axes = axes.flatten() if rows_per_page > 1 else [axes] if cols_per_page == 1 else axes
    
    for idx, stat in enumerate(page_features):
        ax = axes[idx]
        
        feature_name = stat['features']; # print(feature_name)
        values = stat['unique_values']; # print(values)
        counts = stat['frequencies']; # print(counts)
        
        if len(values) > 0:
            empty_color = '#CCCCCC'
            non_empty_values = [v for v in values if str(v).lower() != 'пусто']
            num_non_empty = len(non_empty_values)
            
            colors = []
            palette = sns.color_palette("husl", max(num_non_empty, 1))
            palette_idx = 0
            
            for v in values:
                if str(v).lower() == 'пусто':
                    colors.append(empty_color)
                else:
                    colors.append(palette[palette_idx % len(palette)])
                    palette_idx += 1
            
            bars = ax.bar(range(len(values)), counts, 
                         color=colors, 
                         alpha=0.7, edgecolor='black')
            
            ax.set_xticks(range(len(values)))
            labels = []
            for v in values:
                v_str = str(v)
                if v_str.lower() in ['nan', 'missing']:
                    v_str = 'пусто'
                labels.append(v_str[:25] + '...' if len(v_str) > 25 else v_str)
            ax.set_xticklabels(labels, rotation=45, ha='right', fontsize=8)
            ax.set_ylabel('Частота', fontsize=10)
            
            title = feature_name[:70] + '...' if len(feature_name) > 70 else feature_name
            ax.set_title(title, fontsize=9, fontweight='bold', pad=5)
            
            ax.grid(True, alpha=0.3, axis='y')
            
            for i, count in enumerate(counts):
                ax.text(i, count, str(count), ha='center', va='bottom', fontsize=7)
        else:
            ax.text(0.5, 0.5, 'Нет данных', ha='center', va='center', 
                   transform=ax.transAxes, fontsize=12)
            ax.set_title(feature_name[:70] + '...' if len(feature_name) > 70 else feature_name, 
                        fontsize=9, fontweight='bold')
            ax.axis('off')
    
    for idx in range(len(page_features), graphs_per_page):
        axes[idx].axis('off')
    
    plt.suptitle(f'Страница {page + 1} из {num_pages}', 
                 fontsize=16, fontweight='bold', y=0.995)
    plt.tight_layout(rect=[0, 0, 1, 0.99])
    
    output_file = f"{reports_dir}/{prefix}_all_feats_distrs_{page + 1}.png"
    plt.savefig(output_file, dpi=200, bbox_inches='tight')
    plt.close()
    
    print(output_file)
```
